1.Cİ VERİ SETİ

```python
data=pd.read_csv("/content/heart.csv")
data
```

```python
# One Proportion Test:
#Null hiptotesis : The person's maximum heart rate > 150 olan kişilerin sayısı %60 oranındadır.
#thalach==The person's maximum heart rate
sample_mean = 0.60

stats.ttest_1samp(data["thalach"], sample_mean)
#aalfa degeri p degerinden büyüktür. Null hipotezi ret edilir.
```

```python
# Two Proportions Testing

G1 = data[data["sex"] == 1]["thalach"]
G2 = data[data["sex"] == 0]["thalach"]

stats.ttest_ind(G1, G2)
#pvalue alfa değerinden büyük olduğu için null hipotezi kabul edilir.
#Erkek ve kadın grupları arasında thalach değerinin istatistiksel olarak anlamlı bir fark yoktur.
```

```python
#One Mean Test:
#Null hiptotesis: thalach alanının ortalama degeri 0.8'dır.

sample_mean = 0.8
stats.ttest_1samp(data["thalach"], sample_mean)
##Alfa degeri p degerinden büyüktür.Null hipotezi ret edilir.
```

3.cü Data Seti Vote ile ilgili olan

```python
!wget https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data
column_headers = ['party','handicapped-infants','water-project',
                  'budget','physician-fee-freeze', 'el-salvador-aid',
                  'religious-groups','anti-satellite-ban',
                  'aid-to-contras','mx-missile','immigration',
                  'synfuels', 'education', 'right-to-sue','crime','duty-free',
                  'south-africa']

a= pd.read_csv('house-votes-84.data', names=column_headers)
a= a.replace({"y": 1, "n": 0, "?": np.NaN})
a.head()
```

```python
a.isnull().sum()
```

```python
#One Proportion Testing
#physician-fee-freeze columunudaki nan degerleri  yok ettim.
a_without_nan = a["physician-fee-freeze"].dropna()

# H0: physician-fee-freeze'a yes diyenlerin sayısı %80'dir
sample_mean = 0.8
stats.ttest_1samp(a_without_nan, sample_mean)

#pvalue alfa'dan küçük olduğu için null hipotezini red  ediyoruz ve istatistiksel olarak anlamlı bir fark  vardır.
```

```python
#Two Proportions Test
#Nan degerlerini colum üzerinden attım.
category1 = a[a["party"] == "republican"]["physician-fee-freeze"].dropna()
category2 = a[a["party"] == "democrat"]["physician-fee-freeze"].dropna()

stats.ttest_ind(category1, category2)
#bu teste iki farklı gruba ait aynı ögeleri karşılaştırmak için kullanılır.
#pvalue alfa'dan küçük olduğu için  aralarında anlamlı bir fark vardır. Null hipotezini geçersiz sayılır.
```

```python
#One Mean Test
# physician-fee-freeze columnının ortalaması  0.8'dir
sample_mean = 0.8
stats.ttest_1samp(a_without_nan, sample_mean)
#pvalue değeri alfadan küçük olduğu için aralarında anamlı bir fark vardır  null hipotezi rededilir.
```
